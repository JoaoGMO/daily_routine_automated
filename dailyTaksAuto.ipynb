```python
import pandas as pd   # Importing Pandas.
```

```python
# Selecting the excel file from my desktop, reading the sheet named "Sheet0" and getting the third row as a header. 
# Getting a glance if it's everything ok with a sample of few rows.

excel = pd.read_excel("excel_sample.xlsx", sheet_name = "Sheet0", header = 2)  
excel.head()
```

```python
# Creating a dictionary with assessors as "keys" and their names as "values".

assessors = {}
assessors['A11111'] = "Ass. #1"
assessors['A22222'] = "Ass. #2"
assessors['A33333'] = "Ass. #3"
assessors['A44444'] = "Ass. #4"
assessors['A55555'] = "Ass. #5" 
assessors['A66666'] = "Ass. #6" 

print(assessors)
```

```python
#   Creating a new column with the assessor's name by getting the value in the dictionary (which represent the assessor's name) 
# and linking to the respective key with the map function.

excel["assessor_name"] = excel["Assessor"].map(assessors)
excel.head()  
              
```

```python
#   Last step was to loop over the values of the dictionary, localize each value in the column "ID do Cliente" with "loc",
# send it to a new dictionaty that lists all clients ID in line with it's assessor and print it to generate the Whatsapp msg.  

final = {}
for v in assessors.values():
    final[v] = excel.loc[excel["assessor_name"] == v,"ID do Cliente"].to_list()

for k, v in final.items():
    print(k,":", v) 
    print("")
```

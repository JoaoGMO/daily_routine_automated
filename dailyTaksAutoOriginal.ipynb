```python
import pandas as pd   # Importing Pandas.
```

```python
# Selecting the excel file from my desktop, reading the sheet named "Sheet0" and getting the third row as a header. 
# Getting a glance if it's everything ok with a sample of few rows.

excel = pd.read_excel("excel_sample.xlsx", sheet_name = "Sheet0", header = 2)  
excel.head()
```

```python
# Creating a dictionary with assessors as "keys" and their names as "values".

assessors = {}
assessors['A11111'] = "Ass. #1"
assessors['A22222'] = "Ass. #2"
assessors['A33333'] = "Ass. #3"
assessors['A44444'] = "Ass. #4"
assessors['A55555'] = "Ass. #5" 
assessors['A66666'] = "Ass. #6" 

print(assessors)
```

```python
#   Looping over the excel and the dictionary for identify which Assessor correspond to each value in the dic. The base ideia 
# was to reproduce a VLOOKUP from excel.

excel["assessor_name"] = "NaN"
for r in range(len(excel["Assessor"])):
    for el in range(len(assessors)):
        if list(assessors.keys())[el] == excel["Assessor"][r]:
            excel["assessor_name"][r] = (list(assessors.values()))[el]
            
excel.head()
```

```python
#  Here the goal is to print the name for each assessor and a list of all clients that received the research survey that day.
# To do this a second dictionary was created so the loop-append could be facilitated.

wpp_msg = {}
for el in range(len(assessors)):
    temp_list = []
    for r in range(len(excel)):
        if list(assessors.keys())[el] == excel["Assessor"][r]:
            temp_list.append(excel["ID do Cliente"][r])
    wpp_msg[list(assessors.values())[el]] = temp_list
    

for k, v in wpp_msg.items():
    print(k,":", v) 
    print("")
```
